# 1 . Traversal Operation Code(go through each node and prints its data ) :

#              *  Forword Traversal 

                  1 . start from 1st node prints its data 
                  2 . Move to the second node prints its data 
                  3 . Continue steps 2 untill last node data is printed  
                  
                  
                  
                  Program Explanation :
                  
                  1 . n=self.head
                  2 . print(n.data)
                  3 . n=n.nref

# Backword Traversing 



```python
class Node:
    def __init__(self,data):
        self.data=data
        self.pref=None
        self.nref=None
        
class doublyLL:
    def __init__(self):
        self.head=None
        
        
# Backword traversing program 

    def print_LL_reverse(self):
        if self.head is None:
            print("Linked list is empty")
        
        
        else :
            n=self.head
            while n.nref is not None :
                 n=n.nref
            while n is not None:
                print(n.data)
                n=n.pref
    
```

# 2 . Insertion Operation 



1 . Insertion when Linked List is empty 

        * create a new node 
        * head is pointing to the new node 
        
2 . At begin

        * ckeck linked list is empty or not 
        * point head to the next refrence of new_node 
        * point new node to the previous reference of head 
        * poit new node to the head 
        
        
3 . At End 


4 . After node 
        
        * check the linked list is empty or not 
        * traverse to given node (x)
        * add after x 
        

5 . Before Node 

        * check the linked list is empty or not 
        * traverse to given node (x)
        * add befor x given node

# 3 . Deletion Operation 

Operations :


            1 . Delete from begin 
            
                Coditions :
                   
                   a. LL is empty or not 
                   b. Check DDL contain only one node 
                   c. Rest case 
                 
            2 . Delete from end 
            
                Conditions :
                  
                  a. DDL is empty
                  b. Only one node 
                  c. Rest case
                  
                  
            3 . Delete by value 
            
                Condition :
                
                  a. if dll is empty
                  b. if only one node is present 
                  c. Given node is head
                  d. Given node is last node
                  e. if given node is not present
                  f. rest conditions(nodes at middle position)
                  
                    
            
            

```python
# Insertion when Linked List is empty 


class Node:
    
    def __init__(self,data):
        self.data=data
        self.pref=None
        self.nref=None
        
class doublyLL:
    def __init__(self):
        self.head=None
        

        
# Forword traversing program 

    def print_LL(self):
        if self.head is None:
            print("Linked list is empty")
        
        
        else :
            n=self.head
            while n is not None :
                print(n.data,"---->",end=" ")
                n=n.nref
                
                
# Backword traversing program 

    def print_LL_reverse(self):
        if self.head is None:
            print("Linked list is empty")
        
        
        else :
            n=self.head
            while n.nref is not None :
                 n=n.nref
            while n is not None:
                print(n.data,"----->",end=" ")
                n=n.pref

                
# Insertion when doubly Linked List is empty 

    def insert_empty(self,data):
        
        if self.head is None :
            new_node=Node(data)
            self.head=new_node
        else:
            print("Linked List is not empty !!!")
            
            
# Insertion at the bigining of the doubly linked list 

    def add_bigin(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            new_node.nref=self.head
            self.head.pref=new_node
            self.head=new_node

# Insertion at the end of the doubly linked list 

    def add_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            n=self.head
            while n.nref is not None:
                n=n.nref
            n.nref=new_node
            new_node.pref=n
            
# Insertion after the given node(x)

    def add_after(self,data,x):
        if self.head is None:
            print("Linked list is Empty !!!")
        else:
            n=self.head
            while n is not None :
                if x==n.data:
                    break
                n=n.nref
            if n is None:
                print("Given node is not present in doubly linked list")
            else :
                new_node=Node(data)
                new_node.nref=n.nref
                new_node.pref=n
                if n.nref is not None :
                    
                    n.nref.pref=new_node
                
                n.nref=new_node
                
# Insertion before the given node(x)
    
    
    def add_before(self,x,data):
        if self.head is None :
            print("Linked is Empty ")
        else:
            n=self.head
            while n is not None:
                if x==n.data:
                    break
                n=n.nref
            if n is None:
                print("Given node is not present in DLL")
                
            else:
                new_node=Node(data)
                new_node.nref=n
                new_node.pref=n.pref
                if n.pref is not None:
                    n.pref.nref=new_node
                else:
                    self.head=new_node
                n.pref=new_node

                
# Deletion before the given node. 

    def delete_begin(self):
        if self.head is None :                                   # LL is empty or not 
             print("DLL is empty !!!!")
        elif self.head.nref is None:            # if Check DDL contain only one node 
            print(self.head.data," is deleted")
            self.head=None
        else :                                       
            self.head=self.head.nref                        # Rest case 
            self.head.pref=None

# Deletion at the end of the given node. 

    def del_end(self):
        
        # check DLL is empty or not 
        
        if self.head is None :
            print("DDL is empty ")
            
        # check if contain only one node 
        
        elif self.head.nref is None :
            print(self.head.data," is deleted")
            self.head=None
            
        # rest case 
        
        else:
            n=self.head
            while n.nref is not None:
                n=n.nref
            n.pref.nref=None
                                                                                                                                                             
# delete by value 

    def del_by_val(self,gvn_node):
       
        if self.head is None:                                 #  if dll is empty
            print("DLL is empty")
            
        elif self.head.data==gvn_node and self.head.nref is None:        # if only one node is present 
            self.head=None
            return
            
        
        elif self.head.data==gvn_node:               # Given node is head
            self.head.nref.pref=None
            self.head=self.head.nref
            return
                
           
            
        n=self.head
        while n.nref is not None :
            if n.data==gvn_node:
                break
            n=n.nref
        if n.nref is None:                    
            if n.data==gvn_node:                        # Given node is last node
                 n.pref.nref=None
            else:
                print(f'{gvn_node} is  not present ')            # if given node is not present 
               
        else:                                 
            n.pref.nref=n.nref                  # rest condition       
            n.nref.pref=n.pref
        
        
        
       
            
            
            
                
            
obj=doublyLL()
obj.insert_empty(56)
obj.add_bigin(40)
obj.add_end(60)
obj.add_after("tejas",60)
obj.add_after(569,60)
obj.add_after(490,60)
obj.add_after(101,'tejas')
# obj.delete_begin()
obj.print_LL()
print()
# obj.del_end()
obj.del_by_val(232)
obj.print_LL()
```

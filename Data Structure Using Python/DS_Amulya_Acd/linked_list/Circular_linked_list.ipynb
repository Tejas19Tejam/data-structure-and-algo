# Circular Linked List 


 It is a type of linked list in which every node is conected to each other,in such a way that it form a circle .
 
 1. Circlular link list  is a singly linked 
 2. Circlar link list is a doubliy linked list 
  
  
      Operations :
       
                   1. Insertion 
                   
                           * At the beginig
                           * At the end 
                           * At the middle 
                           
                           
                   2. Deletion
                   
                           * At the beginig
                           * At the end 
                           * At the middle / By value 
                           
                   3. Traversal 
                   
                   
 
 
 

```python
# create node 

class Node:
  def __init__(self,data):
    self.data=data
    self.ref=None

# create Circular LL 

class CirLL:
  def __init__(self):
    self.head=None

# Print Linked List 

  def printLL(self):
    if self.head is None:
      print("Linked list is empty")
    else: 
      head=self.head
      n=self.head
      if n.ref is head:
        print(n.data,end=" -------> ")

      else:
        while n.ref is not head:
          print(n.data,end=" -------> ")
          n=n.ref
        
  def insEmpty(self,data):
    if self.head is None:
      new_node=Node(data)
      self.head=new_node
      self.head.ref=new_node

      



# Driver Code 

if __name__=="__main__":
  obj=CirLL()
  obj.insEmpty(45)
  obj.printLL()



```
